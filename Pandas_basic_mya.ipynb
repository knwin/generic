### slicing ဖြတ်ထုတ်ခြင်း
list slicing နှင့် သဘောတရား အတူတူပင်ဖြစ်သည်။
#### row slicing
`df[start index: end index]`

df[:10] or df[0:10] first ten rows

df[10:] from 11th row to end

df[10:15] from 11th row to 15th

df[-10:] last ten rows

df[:10:2] from start to 10th row with 2 step . i.e even index rows

df[:10:-2] from end to 10th row with 2 step

```python
df[-10::3]
```

#### column slicing
`df[:]['column name'] or df[:]['column name list']`

```python
df_temp = df[:][['SR_Pcode','D_Pcode','TS_Pcode']]
df_temp.head()
```

#### row + column slicing
`df[index start: index end]['column name' or 'column name list']`
ဉပမာ df[10:20][['SR_Pcode','State_Region']] ြဖင့် 'SR_Pcode'နှင့်'State_Region'ကော်လံနှစ်ခုသာပါ သည့် ၁၁ ကြောင်း မှ ၁၉ ကြောင်းအထိဒေတာ များကို လွှာထုတ်နိုင်သည်။

```python
df[10:20][['SR_Pcode','State_Region']]
```

#### selection ပြုလုပ်ခြင်း
`iloc` - index base locating  နှင့် `loc` - locating ဟုနည်း နှစ်နည်းရှိသည်။

iloc သည် index နံပါတ် များ ဖြင့် selection လုပ်ခြင်းဖြစ်သည်။

loc သည် value များ ဖြင့် selection လုပ်ခြင်းဖြစ်သည်။

```python
df.iloc[[0,2,5,]] # row index 0, 2 and 5 will be selected
```

```python
# Kachin ဒေတာ သပ်သပ် select လုပ် ခြင်း
df_kachin = df.loc[df['State_Region']=='Kachin']
df_kachin.head()
```

```python
# Htantabin နှစ်မြို့နယ် အနက် ရန်ကုန်တိုင်းမှ Htantabin ကိုသာ select လုပ် ခြင်း
# condition/criteria ရေးနည်းနှင့် condition/criteria နှစ်ခုကို & ဖြင့် ဆက်ထား သည် ကို သတိပြုပါ
df_ygn_htantabin = df.loc[(df['Township']=='Htantabin')&(df['State_Region']=='Yangon')]
df_ygn_htantabin.head(3)
```
